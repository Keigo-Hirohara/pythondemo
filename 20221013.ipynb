```python
import numpy as np
na = np.array( [[1, 3, 5], [2, 4, 6]] )
nb = np.array( [[1, 4, 7], [10, 13, 16]] )
print("1----------¥n",na + nb)
print("1----------¥n",na + nb)
print("2----------¥n",na * nb)
print("3----------¥n",na.dot(nb.T))
print("4----------¥n",na[1, 2])
print("5----------¥n",na.shape)
 
```

```python
print("6----------¥n",na.reshape(3,2))
print("7----------¥n",na.T)
a = np.array([[1., 3.], [2., 4.]])
print("8----------¥n",np.linalg.inv(a))
```

```python
matrix = np.array([[3,1],[2,4]])
w,v=np.linalg.eig(matrix)
print("9----------¥n",w)
print("10----------¥n",v)
```

```python
import numpy as np
data1= np.random.randint(1, 1000, (1000,1))
print(data1)
```

```python
print('合計: ', np.sum(data1))
print('平均: ', np.mean(data1))
print('分散: ', np.var(data1))
print('標準偏差: ', np.std(data1))
print('中央値: ', np.median(data1))
```

```python
print('最小値: ', np.min(data1))
print('最小値のインデックス: ', np.argmin(data1))

print('最大値: ', np.max(data1))
print('最大値のインデックス: ', np.argmax(data1))
```
